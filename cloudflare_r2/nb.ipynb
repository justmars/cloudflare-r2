```python
from cloudflare_r2 import CloudflareR2Bucket

b = CloudflareR2Bucket(name="prodigy-backup")
```

```python
b.all_items()
```

```python
from pathlib import Path

folder = Path().home() / ".prodigy"
for file in folder.iterdir():
    if file.is_file() and len(file.suffix) > 0:
        f = f"{file.stem}{file.suffix}"
        b.upload(file, f)
```
