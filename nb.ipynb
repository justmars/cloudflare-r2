```python
from src.r2 import CloudflareR2Bucket

b = CloudflareR2Bucket(name="law")
```

```python
res = b.get(key="main.db")
chunks = res["Body"].iter_chunks()
```

```python
with open("main.db", "wb") as f:
    for chunk in chunks:
        f.write(chunk)
```

```python
from pathlib import Path

folder = Path().home() / ".prodigy"
for file in folder.iterdir():
    if file.is_file() and len(file.suffix) > 0:
        f = f"{file.stem}{file.suffix}"
        b.upload(file, f)
```
